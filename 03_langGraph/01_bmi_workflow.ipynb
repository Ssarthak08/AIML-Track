```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
# Define states
class BMIstate(TypedDict):
    weight_kg: float
    height_m: float
    bmi: float
    bmi_category: str
```

```python
def calculate_bmi(state: BMIstate) -> BMIstate:
    weight = state['weight_kg']
    height = state['height_m']
    bmi = weight / (height ** 2)
    state['bmi'] = round(bmi, 2)
    return state
```

```python
def bmi_label(state: BMIstate) -> BMIstate:
    bmi = state['bmi']
    if bmi < 18.5:
        state['bmi_category'] = 'Underweight'
    elif 18.5 <= bmi < 25:
        state['bmi_category'] = 'Normal weight'
    elif 25 <= bmi < 30:
        state['bmi_category'] = 'Overweight'
    else:
        state['bmi_category'] = 'Obese'
    return state
```

```python
# define graph
graph = StateGraph(BMIstate)
# add nodes

graph.add_node('calculate_bmi', calculate_bmi)
graph.add_node('bmi_label', bmi_label)

# add edges
graph.add_edge(START, 'calculate_bmi')
graph.add_edge('calculate_bmi', 'bmi_label')
graph.add_edge('bmi_label', END)
 
# compute graph
workflow = graph.compile()
```

```python
initial_state = BMIstate(weight_kg=70, height_m=1.75, bmi=0, bmi_category="")

final_state = workflow.invoke(initial_state)

print(final_state)
```
