```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
class BatsmanState(TypedDict):
    runs: int
    balls: int
    fours: int
    sixes: int

    sr: float
    bpb: float
    boundary_percentage: float
    summary: str
    
```

```python
def calculate_sr(state: BatsmanState) -> BatsmanState:

    state['sr'] = (state['runs'] / state['balls']) * 100
    return {'sr': state['sr']}
```

```python
def calculate_bpb(state: BatsmanState) -> BatsmanState:

    state['bpb'] = state['balls'] / (state['fours'] + state['sixes'])
    return {'bpb': state['bpb']}
```

```python
def calculate_boundary_percentage(state: BatsmanState) -> BatsmanState:

    state['boundary_percentage'] = ((state['fours'] + state['sixes']) / state['runs']) * 100
    return {'boundary_percentage': state['boundary_percentage']}
```

```python
def summary_score(state: BatsmanState) -> BatsmanState:
    summary = f'''
    Runs: {state['runs']}
    Balls: {state['balls']}
    Fours: {state['fours']}
    Sixes: {state['sixes']}
    Strike Rate: {state['sr']}
    Balls per Boundary: {state['bpb']}
    Boundary Percentage: {state['boundary_percentage']}
    '''
    state['summary'] = summary
    return {'summary': state['summary']
    }  
```

```python
graph = StateGraph(BatsmanState)

graph.add_node('calculate_sr',calculate_sr)
graph.add_node('calculate_bpb',calculate_bpb)
graph.add_node('calculate_boundary_percentage',calculate_boundary_percentage)
graph.add_node('summary',summary_score)

graph.add_edge(START, 'calculate_sr')
graph.add_edge(START, 'calculate_bpb')  
graph.add_edge(START, 'calculate_boundary_percentage')
graph.add_edge('calculate_sr', 'summary')
graph.add_edge('calculate_bpb', 'summary')
graph.add_edge('calculate_boundary_percentage', 'summary')
graph.add_edge('summary', END)
```

```python
workflow = graph.compile()
```

```python
input_state = BatsmanState(
    runs=100,
    balls=60,
    fours=10,
    sixes=5,
    sr=0.0,
    bpb=0.0,
    boundary_percentage=0.0,
    summary=''
)
workflow.invoke(input_state)
```
