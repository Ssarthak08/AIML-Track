```python
from langchain_text_splitters import HTMLHeaderTextSplitter 
html_string = '''

<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8" />
  <title>User Account Documentation</title>
</head>
<body>

  <h1>User Account</h1>
  <p>This document explains user account details and settings.</p>

  <h2>Profile Information</h2>
  <p>Basic information related to the user profile.</p>

  <h3>Personal Details</h3>
  <p>Name, email address, and contact number of the user.</p>

  <h3>Account Status</h3>
  <p>Shows whether the account is active, suspended, or closed.</p>

  <h2>Security Settings</h2>
  <p>Settings related to authentication and authorization.</p>

  <h3>Password Policy</h3>
  <p>Password must be at least 8 characters long.</p>

  <h3>Two-Factor Authentication</h3>
  <p>Enable 2FA for enhanced account security.</p>

</body>
</html>

'''

headers_to_split_on = [
    ("h1","Header 1"),
    ("h2","Header 2"),
    ("h3","Header 3")

]
html_splitter=HTMLHeaderTextSplitter(headers_to_split_on)
html_splitter_splits=html_splitter.split_text(html_string)  
html_splitter_splits
```
