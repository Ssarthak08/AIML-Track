```python
user_age = int(input("Please enter your age: "))
```

```python
print(f"This is your age: {user_age}")
```

```python
if user_age < 18:
    print("You are a minor.")
elif user_age < 50:
    print("You are an adult.")
elif user_age < 100:
    print("You are a senior.")
else:
    print("You are a centenarian.")
```
