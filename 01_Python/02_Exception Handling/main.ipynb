### try and except block is used when we expect that there is some code that can go wrong in the runtime. So try and except block is used when we want to open an external resource.

### try: (error causing code), except: (instead what should be done)

```python
with open("file.txt", "w") as f:
    f.write("Hello from Python\nThis is AIML file handling")
```

```python
# catching specific errors 

try : 
    f = open('file1.txt','r')
except :
    print('File not found')
```

```python
try : 
    f = open('file1.txt','r')  # now file1 doesn't exist 
except Exception as e :
    print(e.with_traceback)
```
